### Ejemplos de if, if else, if elif

# Ejemplo de if

```python
# 1. Ejemplo de if simple con operadores aritméticos y relacionales
#Este código verifica si el cuadrado de un número supera un límite determinado.
numero = 12
# Se utiliza el operador aritmético de potencia (**) y el relacional (>)
if numero ** 2 > 100:
    print("El cuadrado del número es mayor a 100.")
```

```python
# 2. Ejemplo de if-else con operadores relacionales y lógicos
# Este ejemplo determina si una persona puede acceder a un descuento basado en su edad y su estatus de estudiante.
edad = 17
# Se utilizan operadores relacionales (>=) y el operador lógico (and)
if edad < 18:
    print("Tienes derecho a un descuento.")
else:
    print("Debes pagar la tarifa completa.")
```

```python
# 3. Ejemplo de if-elif-else con operadores de módulo y lógicos
# Este código categoriza un número según su relación con otros valores.
n = int(input("Ingresa un número: "))

# Uso de operador de módulo (%) y relacional (==)
if n % 2 == 0:
    print("El número es par.")
# Uso de módulo, relacional y lógico (and)
elif n % 5 == 0 and n > 10:
    print("El número es múltiplo de 5 y mayor a 10.")
else:
    print("El número no cumple las condiciones anteriores.")
```

```python
# 4. Ejemplo combinado (Operadores logicos y relacionales en un rango)
# Verifica si una calificacion esta en un rango valido para aprobacion.
nota = 0
if nota >= 0 and nota <= 10:
  if nota >= 6:
    print("Aprobado")
  else:
    print("Reprobado")
else:
  print("Nota fuera de rango")
```
